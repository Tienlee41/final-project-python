###  Handing null value

#### Drop columns have full null-values, duplicated values

```python
def drop_null_dup(df):
    df.dropna(axis=1,how='all',inplace=True)
    df.drop_duplicates(inplace=True)
drop_null_dup(starbucks_df)
drop_null_dup(highlands_df)
drop_null_dup(coffee_house_df)
```

- Shared text column and link column

    trong 2 cot shared text,post text,text,links link,with se co nhung cot co data va k co data


```python
def fill_text(df):
    df['text'].fillna('no text',inplace=True)
    df['shared_text'].fillna('no text',inplace=True)
    df['post_text'].fillna('no text',inplace=True)
    df['link'].fillna('no link',inplace=True)
    if 'with' in df.columns:
        df['with'].fillna('no with',inplace = True)
fill_text(starbucks_df)
fill_text(highlands_df)
fill_text(coffee_house_df)
```

#### Transform reactions columns

```python
def change_reaction(df):
    df['reactions'].to_dict()
    new_col_list = {'like':'thích',
                    'love':'yêu thích',
                    'care':'thương thương',
                    'haha':'haha',
                    'wow':'wow',
                    'sad':'buồn',
                    'angry':'phẫn nộ'}
    for col in new_col_list:
        df[col] = 0
    for i in range(100):
        for col in new_col_list:
            if new_col_list[col] in df['reactions'][i]:
                reac = dict(eval(df['reactions'][i]))
                df[col][i] = reac[new_col_list[col]]
                
    return df
```

```python
starbucks_df = change_reaction(starbucks_df)
highlands_df = change_reaction(highlands_df)
coffee_house_df = change_reaction(coffee_house_df)
```

## Convert df after clean to csv file

```python
def convert_to_csv(df):
    path= 'Data/Cleaning_data/' + df.name + "_cleaning_data" + ".csv"
    df.to_csv(path,index = False)
    print(path)
convert_to_csv(starbucks_df)
convert_to_csv(highlands_df)
convert_to_csv(coffee_house_df)
```
