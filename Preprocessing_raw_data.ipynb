## Step 1 : Data cleaning

In this step, I will handing incorrect, incomplete, inconsistent and missing values

###  Handing null value

#### Drop columns have full null-values, duplicated values

```python
def drop_null_dup(df):
    df.dropna(axis=1,how='all',inplace=True)
    df.drop_duplicates(inplace=True)
drop_null_dup(starbucks_df)
drop_null_dup(highlands_df)
drop_null_dup(coffee_house_df)
```

#### Filling the null-values

```python
def num_of_null_values(df):
    col = df.columns
    for i in range(len(col)):
        if (df[col[i]].isna().sum() + (df[col[i]] == '[]').sum()) > 0:
            print(col[i],(df[col[i]].isna().sum() + (df[col[i]] == '[]').sum()))
num_of_null_values(starbucks_df)
```

```python
def fill_text(df):
    df['text'].fillna('no text')
    df['shared_text'].fillna('no text')
    df['post_text'].fillna('no text')
    df['link'].fillna('no link')
    if 'with' in df.columns:
        df['with'].fillna('no with')
fill_text(starbucks_df)
fill_text(highlands_df)
fill_text(coffee_house_df)
```

## Step 4 : Data Transformation

#### Transform reactions columns

```python
def trans_reac(df):
    reactions_df = df[['reactions']]
    reactions_df['reactions'] = reactions_df['reactions'].apply(lambda x : dict(eval(x)))
    reactions_df = reactions_df['reactions'].apply(pd.Series)
    reactions_df.fillna(0,inplace = True )
    ndf = df.join(reactions_df)
    ndf = df.drop('reactions', axis=1)
    ndf.name = df.name
    return ndf
```

```python
starbucks_df = trans_reac(starbucks_df)
highlands_df = trans_reac(highlands_df)
coffee_house_df = trans_reac(coffee_house_df)
```

## Convert df after clean to csv file

```python
def convert_to_csv(df):
    path= 'Data/Cleaning_data/' + df.name + "_cleaning_data" + ".csv"
    df.to_csv(path,index = False)
    print(path)
convert_to_csv(starbucks_df)
convert_to_csv(highlands_df)
convert_to_csv(coffee_house_df)
```
